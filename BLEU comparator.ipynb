```python
import pandas as pd

# Assuming complete_seqs, complete_seqs_scores, and references are already populated

# Create a dictionary to hold the data
data = {
    'BLEU1': obj1,
    'BLEU2': obj2,
    'BLEU3': obj3,
    'BLEU4': obj4,
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
avg_df_1 = df['BLEU1'].rolling(window=50).mean()
avg_df_2 = df['BLEU2'].rolling(window=50).mean()
avg_df_3 = df['BLEU3'].rolling(window=50).mean()
avg_df_4 = df['BLEU4'].rolling(window=50).mean()
```

```python
df['avg_df_1'] = avg_df_1
df['avg_df_2'] = avg_df_2
df['avg_df_3'] = avg_df_3
df['avg_df_4'] = avg_df_4
```

```python
df
```

```python
import pandas as pd

# Assuming you have a DataFrame named 'df'
# Create an empty DataFrame to store the selected records
selected_df = pd.DataFrame(columns=df.columns)

# Filter records that are multiples of 5
for index, row in df.iterrows():
    if index % 50 == 0:  # Check if the index is a multiple of 5
        selected_df = selected_df.append(row)
```
